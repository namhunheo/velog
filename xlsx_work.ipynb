```python
import glob
import pandas as pd
```

```python
# glob을 이용해 매출 파일 목록 불러오기
files = glob.glob("sample_data/월별매출내역_*.xlsx")
files
```

```python
total = pd.DataFrame()

for file in files:
    # 상단의 6줄은 제목행으로 건너뜀
    df = pd.read_excel(file, skiprows=range(6)).iloc[:, 1:]
    total = pd.concat([total, df], ignore_index=True)

print(len(total))
total.head()
```

```python
price = pd.read_excel("sample_data/상품가격.xlsx")
price = price.transpose()  # 행/열 전치(행과 열 반대로)
price.head()
```

```python
price_dict = price.iloc[1:, 0].to_dict()
price_dict
```

```python
total['가격'] = total['품명'].map(price_dict)
total.head()
```

```python
total['매출금액'] = total['수량'] * total['가격']
total.head()
```

```python
total['일자'] = pd.to_datetime(total['일자'])
total['월'] = total['일자'].dt.month
total[['일자', '월']].head()
```

```python
result = total.groupby(['월', '지점'])['매출금액'].sum().reset_index()
result
```

```python
ew = pd.ExcelWriter('result.xlsx', engine='xlsxwriter')
result.to_excel(ew, index=False)
ew.close()
```
